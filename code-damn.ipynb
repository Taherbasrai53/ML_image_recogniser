```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# New Section

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
'''
0 => T-shirt/top 
1 => Trouser 
2 => Pullover 
3 => Dress 
4 => Coat 
5 => Sandal 
6 => Shirt 
7 => Sneaker 
8 => Bag 
9 => Ankle boot '''
```

```python
plt.imshow(X_train[0], cmap ="Greys")
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_train[i],cmap="Greys")
  plt.axis('off')
  plt.title(class_labels[y_train[i]]+"="+str(y_train[i]), fontsize=20)

  '''
0 => T-shirt/top 
1 => Trouser 
2 => Pullover 
3 => Dress 
4 => Coat 
5 => Sandal 
6 => Shirt 
7 => Sneaker 
8 => Bag 
9 => Ankle boot '''
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_test[0]
```

### build nueral network

```python
model = keras.models.Sequential([
                         keras.layers.Flatten(input_shape=[28,28]),
                         keras.layers.Dense(units=32, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1)
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred[0].round(2)
```

```python
np.argmax(y_pred[0].round(2))
```

```python
plt.figure(figsize=(16, 16))
for i in range(25):
   plt.subplot(5, 5, i+1)
   plt.imshow(X_test[i],cmap="Greys")
   plt.axis('off')
   plt.title("Displayed ={}".format(class_labels[np.argmax(y_pred[i])]))
```

```python
plt.figure(figsize=(16, 16))
plt.subplot(10, 10, 99)
plt.imshow(X_test[99],cmap="Greys")
plt.axis('off')
plt.title("Displayed ={}".format(class_labels[np.argmax(y_pred[99])]))
```
